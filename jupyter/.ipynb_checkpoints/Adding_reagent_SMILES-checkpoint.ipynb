## Transformation of Reagents to SMILES by Pubchempy and Cirpy

pubchempy and cirpy are python packages which can be used to translate reagent names into SMILES codes. Once we have the reagents as SMILES, we can add them to the reactants side of the reaction equation. We can then get a more accurate representation of the reactions when computing reaction fingerprints. Using reaction fingerprints we can compare the reactions in chemical space and perform clustering analysis to begin to classify the reactions.

```python
# Import the useful reactions we extracted from our data processing escapades
data = pd.read_csv(pwd+'/data/useful_reactions.csv')

# Set logfile filepath
logfile = pwd+'/logs/adding_reagent_SMILES.txt'
```

### Cirpy

```python
# To add in reagents as SMILES first convert as many reagents as possible using cirpy

# Isolate unique reagents from one step reactions
reagents = [i.split('; ') for i in data['Reagent'].to_list()]
unique_reagents = set(list(itertools.chain(*reagents)))

# Logfile entry - total # items converted by cirpy in raw dataset
addToFile(logfile, ('Total number of unique reagents to convert to SMILES: {}\n'
                    .format(len(unique_reagents))))

# Store the unsuccessful names in a list for 
# later processing by  pubchempy
converted_smiles = []
converted_name = []
unsuccessful_reagents = []
for i in unique_reagents:
    # Call cirpy.resolve to change from name -> SMILES
    smiles = cirpy.resolve(i, 'smiles', ['name_by_cir','name_by_opsin'])
    if smiles != None:
        converted_name.append(i)
        converted_smiles.append(smiles)
    else:
        unsuccessful_reagents.append(i)
            
# Logfile entry - total # items converted by cirpy in raw dataset
addToFile(logfile, ('Total number of reagents converted by cirpy: {}\n'
                    .format(len(converted_smiles))))
```

### Pubchempy

```python
# pubchempy processing of the unsuccessful cirpy list


# Get the pubchempy id's for each reagent if it exists
compound_ids = []
for i in unsuccessful_reagents:
    results = pcp.get_compounds(i, 'name')
    compound_ids.append(results)
    
# Make the compound ids into a list of strings so that the numeric part can be exctracted
string_ids_list = []
for sublist in compound_ids:
    string_ids = []
    for compound_id in sublist:
        string_ids.append(str(compound_id))
    string_ids_list.append(string_ids)

# Extract the numeric portion of the strings for conversion to smiles
cp_id_nums_list = []
for sublist in string_ids_list:
    cp_id_nums = []
    for str_id in sublist:
        cp_id_nums.append(''.join([s for s in str_id if s.isdigit()]))
    cp_id_nums_list.append(cp_id_nums)
    
# Get the cannonical smiles representation of the compound and store in a list
smiles_list = []
for sublist in cp_id_nums_list:
    smiles = []
    for id_num in sublist:
        c = pcp.Compound.from_cid(id_num)
        smiles.append(c.canonical_smiles)
    smiles_list.append(smiles)

# Convert smiles list to strings and take only first representation if multiple available
str_smiles_list = []
for sublist in smiles_list:
    if len(sublist) >= 1:
        sublist = str(sublist[0])
        str_smiles_list.append(sublist)
    else:
        str_smiles_list.append('NA')
        
# Dataframe to store name and smiles representation
reags_with_smiles_df = pd.DataFrame(unsuccessful_reagents, columns=['Reagent'])
reags_with_smiles_df['Smiles'] = str_smiles_list

# Remove the reagents which couldn't be processed
mask = reags_with_smiles_df['Smiles'] != 'NA'
successful_pubchempy = reags_with_smiles_df[mask]

# Create lists of converted reagents converted by both methods
successful_names_all = successful_pubchempy['Reagent'].to_list() + converted_name
successful_smiles_all = successful_pubchempy['Smiles'].to_list() + converted_smiles

# Store in DataFrame
converted_df = pd.DataFrame(successful_names_all, columns=['Reagent Name'])
converted_df['SMILES'] = successful_smiles_all
converted_df.to_csv(pwd+'/data/reagents_to_smiles_df.csv')
```

```python
# Logfile entry - total # items converted by pubchempy in raw dataset
addToFile(logfile, ('Total number of reagents converted by pubchempy: {}\n'
                    .format(len(converted_df)-len(converted_smiles))))
```

```python
# Get unsuccessful reagents
unsuccessful_reagents = [i for i in unique_reagents if i not in converted_df['Reagent Name'].to_list()]

# Logfile entry - number of reagents which couldn't be converted
addToFile(logfile, ('Total number of unsuccessful reagents: {}\n'
                    .format(len(unsuccessful_reagents))))
```

## Joining the reagents onto the reactants

start by adding the reagents to the correct reactions, then separate out the reaction SMILES into reactants and products, join the reagents to the reactants side and put the reaction string back together

```python
# Extract the smiles and names into lists with common index
names = reagents_as_SMILES['Reagent Name'].to_list()
smiles = reagents_as_SMILES['SMILES'].to_list()

# Get the reagents from the master dataset
reagents = [i.split('; ') for i in data['Reagent'].to_list()]

# cross reference names to smiles and replace names with corresponding smiles strings
# store in converted_to_smiles which is a list of smiles strings and names which
# were not converted
converted_to_smiles = []
for sublist in reagents:
    sublist2 = []
    for name in sublist:
        if name in names:
            sublist2.append(smiles[names.index(name)])
        else:
            sublist2.append(name)
    converted_to_smiles.append(sublist2)
    
# extract the reagents which were not converted into weird_reagents
# store the rest of the converted reagents in add_to_reactants
weird_reagents = []
add_to_reactants = []
for sublist in converted_to_smiles:
    sublist2 = []
    sublist3 = []
    for item in sublist:
        if item not in smiles:
            sublist2.append(item)
        else:
            sublist3.append(item)
    weird_reagents.append(sublist2)
    add_to_reactants.append(sublist3)

# Add converted reagents and non converted reagents to master df
data['Reagents as SMILES'] = ['.'.join(i) for i in add_to_reactants]
data['Reagents Not Converted'] = [', '.join(i) for i in weird_reagents]
```

```python
# Logfile entry - number of reagents which couldn't be converted in the master dataset
addToFile(logfile, ('Total number of unsuccessful reagents in the useful reactions dataset: {}\n'
                    .format(len(set(data['Reagents Not Converted'].to_list())))))

# Logfile entry - number of reactions with reagents that couldn't be converted
addToFile(logfile, ('Total number of reactions with unsuccessful reagents: {}\n'
                    .format(len(data) - 
                           (len(data[data['Reagents Not Converted'].isin(['None'])]) +
                            len(data[data['Reagents Not Converted'].isin([''])])))))
```

```python
# Now we need to make the new reaction SMILES Strings

# The reactions without reagents to add can be left unchanged
without_reagents = data[data['Reagents as SMILES'].isin([''])]

# Add in a new column to store the reaction SMILES for concatenation later
without_reagents['Reactants.Reagents>>Products']= without_reagents['Reaction']

# Logfile entry - number of reactions which don't have additional reagents
addToFile(logfile, ('Number of reactions which dont have additional reagents: {}\n'
                    .format(len(without_reagents))))

# The ones with reagents to add need to be deconstructed then reassembled
with_reagents = data[~data['Reagents as SMILES'].isin([''])]

# Logfile entry - number of reactions which don't have additional reagents
addToFile(logfile, ('Number of reactions which have additional reagents: {}\n'
                    .format(len(with_reagents))))
```

```python
# Split reaction SMILES into reactants and products
with_reagents['Reactants'] = with_reagents['Reaction'].str.rsplit(">>", n=1, expand=True)[0]
with_reagents['Products'] = with_reagents['Reaction'].str.rsplit(">>", n=1, expand=True)[1]

# Combine the reactants and reagents columns
with_reagents['Reactants + Reagents'] = with_reagents[
                                                ['Reactants', 'Reagents as SMILES']].agg('.'.join, axis=1)

# Combine new reactant column with products column
with_reagents['Reactants.Reagents>>Products'] = with_reagents[
                                                ['Reactants + Reagents', 'Products']].agg('>>'.join, axis=1)
```

```python
# Now we need to unify the two reaction smiles columns
# and stick them back into the original dataframe
reagents_added = pd.concat([without_reagents, with_reagents])

# Finally we need to save the resulting dataframe as a csv file for further use
reagents_added.to_csv(pwd+'/data/reagents_added.csv', index=False)
```

## Creating a second version of Reaction SMILES for reaction core generation
## with a view to reaction classification

To generate a reaction core i.e. the atoms/bonds involved in the reaction we need to map the reactant atoms into
the products, via the process of atom-atom mapping. The KNIME atom-atom mapping tool requires that the reaction is
represented in a slightly different way than required to generate the differential reaction fingerprints. The
difference is that instead of putting the reagent SMILES in with the reactants and separating reactants from products
with '>>', the reagents are inserted in between reactants and products with one '>' after reactants, and one '>' after
reagents. Making the reaction SMILES template = reactants>reagents>products

```python
# Re import the reagents_added dataframe from it's csv file stored in the data folder
# this avoids us having to run the previous code boxes
data2 = pd.read_csv(pwd+'/data/reagents_added.csv')
```

```python
# We dont need to change the reaction SMILES for those without reagents
# So just keep the original reaction SMILES
without_reagents2 = data2[data2['Reagents as SMILES'].isna()]

without_reagents2['Reactants>Reagents>Products']= without_reagents2['Reaction']
```

```python
# Now we get all the reactions which do have a reagent
with_reagents2 = data2[~data2['Reagents as SMILES'].isna()]

# And create the new reaction SMILES with format reactants>reagents>products

# Combine the reactants and reagents columns
with_reagents2['Reactants>Reagents'] = with_reagents2[
                                                ['Reactants', 'Reagents as SMILES']].agg('>'.join, axis=1)

# Combine new reactant column with products column
with_reagents2['Reactants>Reagents>Products'] = with_reagents2[
                                                ['Reactants>Reagents', 'Products']].agg('>'.join, axis=1)
```

```python
# Now we can join the two dataframes together and save as a csv file
reagents_added = pd.concat([without_reagents2, with_reagents2])

# This csv file will be transferred to KNIME for atom mapping and reaction core generation
reagents_added.to_csv(pwd+'/data/reactants>reagents>products.csv', index=False)
```
